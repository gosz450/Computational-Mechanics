# Newton's method for a function of one variable $f(x) = x^2-2 = 0$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')
```

```python
# define f(x)
f = x**2-2
f
```

```python
# define f'(x)
fp = sp.diff(f,x)
```

```python
fp
```

```python
# Define initial guess and number of iterations
xi = 1
niter = 6
# Storage
Ri_vals = []           # residual history (floats)
xi_vals = []           # iterate history (optional)
```

```python
# Newton's iterations
for i in range(niter):
    fi = f.subs(x,xi)
    fpi = fp.subs(x,xi)
    dxi = -fi/fpi
    Ri = sp.Abs(fi)
    # compute the residual R_i = |f(x_i)|
    # store as float for plotting
    Ri_vals.append(float(sp.N(Ri)))
    xi_vals.append(float(sp.N(xi)))
    # Newton update
    xi = xi + dxi
```

### Plot the residual $R^i$ versus the number of iterations

```python
# --- convert to numpy array (optional) ---
Ri_arr = np.array(Ri_vals, dtype=float)

# --- plot residual vs iteration ---
plt.figure()
plt.plot(range(1, niter+1), Ri_arr, marker='o')
# For wide dynamic range, semilogy is often clearer:
# plt.semilogy(range(1, niter+1), Ri_arr, marker='o')

plt.xlabel('Iteration i')
plt.ylabel('Residual $R^i$')
plt.title('Newton Residual History')
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
