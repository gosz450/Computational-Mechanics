```python
# Import Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sympy import symbols, Matrix, solve
```

### Create a simple NumPy array

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Plot of the sine function')
plt.show()
```

### Create and manipulate a matrix
This command symbolically solves the system $\mathbf{A}\mathbf{x} = \mathbf{b}$ and returns the solution vector.

```python
A = Matrix([[2, 1], [1, 3]])
b = Matrix([5, 7])
x = A.LUsolve(b)
x
```

### The for loop

```python
for i in range(1, 6):
    print(f"Square of {i} is {i**2}")
```

### Define and call a function
Note: functions are typically placed at the top of the Jupyter notebook after importing packages.

```python
def area_rectangle(a, b):
    return a * b
print(area_rectangle(3, 4))
```
